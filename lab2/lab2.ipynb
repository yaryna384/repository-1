```python
B = randint(0, 10)
if B > 5:
    print(f"B={B} — число більше 5")
elif B == 5:
    print("B дорівнює 5")
else:
    print(f"B={B} — число менше 5")
```

```python
C = randint(0, 20)
if C % 2 == 0:
    if C > 10:
        print(f"C={C} — парне і більше 10")
    else:
        print(f"C={C} — парне і менше або рівне 10")
else:
    print(f"C={C} — непарне")
```

```python
# Звичайна функція
def full_name(first_name, last_name):
    return first_name, last_name

# лямбда-функція
greet = lambda first, last: f"Привіт, {first} {last}! Ласкаво просимо в Python."

print("Звичайна функція:", full_name)
print("Lambda-функція:", greet)
print(greet("Марія", "Іваненко"))
print(greet(*full_name("Олег", "Петренко")))
```

```python
# Змінні різних типів
string_var = "Привіт, Python!"
int_var = 10
float_var = 3.14
list_var = [1, 2, "три", 4.0]
tuple_var = (1, "два", 3.0)
dict_var = {"ім'я": "Аліса", "вік": 25}
set_var = {1, 2, 3, 3}  # дублі видаляються

print(string_var)
print(int_var, float_var)
print(list_var)
print(tuple_var)
print(dict_var)
print(set_var)
```

```python
# Цикл for
for i in range(3):
    print(f"Цикл for: {i}")

# Цикл while
x = 0
while x < 3:
    print(f"Цикл while: {x}")
    x += 1

# Розгалуження if
age = 20
if age >= 18:
    print("Дорослий")
else:
    print("Дитина")
```
